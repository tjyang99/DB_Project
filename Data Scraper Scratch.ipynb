```python
import requests 
from bs4 import BeautifulSoup

import time
import random as rand 

import pandas as pd

pokemon_review_dict = {'name':[], 'date':[], 'rating':[], 'review':[]}

for page in range(0,30): #Remember to update the number of pages 
    url = 'https://www.metacritic.com/game/switch/pokemon-sword/user-reviews?sort-by=most-helpful&num_items=100&page='+str(page)
    user_agent = {'User-agent': 'Mozilla/5.0'}
    response  = requests.get(url, headers = user_agent)
    time.sleep(rand.randint(3,30)) 
    soup = BeautifulSoup(response.text, 'html.parser')
    for review in soup.find_all('div', class_='review_content'):
        if review.find('div', class_='name') == None:
                       break 
        pokemon_review_dict['name'].append(review.find('div', class_='name').find('a').text)
        pokemon_review_dict['date'].append(review.find('div', class_='date').text)
        pokemon_review_dict['rating'].append(review.find('div', class_='review_grade').find_all('div')[0].text)
        if review.find('span', class_='blurb blurb_expanded'):
            pokemon_review_dict['review'].append(review.find('span', class_='blurb blurb_expanded').text)
        else:
            pokemon_review_dict['review'].append(review.find('div', class_='review_body').find('span').text)

sword_reviews = pd.DataFrame(pokemon_review_dict)
```

```python
player_count_dict = {'name':[], '24-count':[], 'all-time-count':[]}
appid = 1240440
url = 'https://steamcharts.com/app/'+str(appid)
user_agent = {'User-agent': 'Mozilla/5.0'}
response  = requests.get(url, headers = user_agent)
time.sleep(rand.randint(3,30)) 
soup = BeautifulSoup(response.text, 'html.parser')
for review in soup.find_all('div', id='content-wrapper'):
    if review.find('h1', id='app-title') == None:
        print("broke")
        break 
    player_count_dict['name'].append(review.find('h1', id='app-title').find('a').text)
    player_count_dict['24-count'].append(int(review.find_all('div', class_='app-stat')[1].find('span', class_='num').text))
    player_count_dict['all-time-count'].append(int(review.find_all('div', class_='app-stat')[2].find('span', class_='num').text))
```

```python
pokemon_review_dict_short = {'name':[], 'rating':[], 'count':[]}

url = 'https://www.metacritic.com/game/switch/pokemon-sword/user-reviews?sort-by=most-helpful&num_items=100&page=0'
user_agent = {'User-agent': 'Mozilla/5.0'}
response  = requests.get(url, headers = user_agent)
time.sleep(rand.randint(3,30)) 
soup = BeautifulSoup(response.text, 'html.parser')
for review in soup.find_all('div', id='main_content'):
    pokemon_review_dict_short['name'].append(review.find('div', class_='product_title').find('h1').text)
    pokemon_review_dict_short['rating'].append(review.find('div', class_='userscore_wrap feature_userscore').find_all('div')[1].text)
    pokemon_review_dict_short['count'].append(review.find('div', class_='summary').find_all('strong')[0].text)
```

```python
halo_review_dict_short = {'name':[], 'rating':[], 'count':[]}

url = 'https://www.metacritic.com/game/playstation-5/call-of-duty-modern-warfare-ii/user-reviews'
user_agent = {'User-agent': 'Mozilla/5.0'}
response  = requests.get(url, headers = user_agent)
time.sleep(rand.randint(3,30)) 
soup = BeautifulSoup(response.text, 'html.parser')
for review in soup.find_all('div', id='main_content'):
    halo_review_dict_short['name'].append(review.find('div', class_='product_title').find('h1').text)
    halo_review_dict_short['rating'].append(float(review.find('div', class_='userscore_wrap feature_userscore').find_all('div')[1].text))
    count_string = review.find('div', class_='summary').find_all('strong')[0].text
    count = int(count_string.split()[0])
    halo_review_dict_short['count'].append(count)
```
